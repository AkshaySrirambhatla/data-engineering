```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("Day1").getOrCreate()
spark
```

```python
data=[
    (1,"Akshay",1000),
    (2,"Kumar",2000),
    (3,"Sri",3000)
]
df=spark.createDataFrame(data,["Id","Name","Salary"])
df.show()
```

```python
df.select("Name","Salary").show()
df.withColumn("Salary_Hike",df.Salary*1.1).show()
df.filter(df.Salary>2000).show()
```

```python
df.filter(df.Id==3).select("Salary")*1.1
```
